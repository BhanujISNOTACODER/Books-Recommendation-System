```python
books.Description = books.Description.apply(lambda X:X.split())
```

```python
from ast import literal_eval

books.Genres = books.Genres.apply(lambda X:literal_eval(X))
```

```python
books["tags"] = books.Author + books.Genres + books.Description
```

```python
def convert(l1):
  L=[]
  for i in l1:
    L.append(i.replace(" ",""))
  return L

books.tags = books.tags.apply(convert)
```

```python
books = books[["title","URL","tags"]]
```

```python
books.tags=books.tags.apply(lambda X:[i.replace(" ","") for i in X])
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

def StemTags(l1):
  L=[]
  for i in l1:
    L.append(ps.stem(i))
  return L

books.tags = books.tags.apply(StemTags)
```

```python
books
```

```python
books["tags"] = books["tags"].apply(lambda X:" ".join(X))
```

```python
books
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000,stop_words="english")
```

```python
vectors = cv.fit_transform(books['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
start = 0

# Add column with incremental values from 30 onwards
books['Ind'] = range(start, start + books.shape[0])

books
```

```python
def recommend(book):
  book_ind = books[books["title"]==book].Ind
  mentioned_book = similarity[book_ind][0]
  similar_books = sorted(list(enumerate(mentioned_book)),key=lambda X:X[1],reverse=True)[1:6]
  L=[]
  for i in similar_books:
    L.append(books.iloc[i[0]].title)
  return L
```

```python
import pickle
pickle.dump(similarity,open('similarity.pkl','wb'))
```
